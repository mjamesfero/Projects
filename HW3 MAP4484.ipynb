```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt

#1a
def dF(x, y):
    return 5*x+10*y, -x-y

def eigen(x,a,b):
    m = a/b
    return m*x

X, Y = np.meshgrid(np.linspace(-3.0, 3.0, 30), np.linspace(-3.0, 3.0, 30))
dX, dY = dF(X, Y)

plt.streamplot(X, Y, dX, dY, color='r', linewidth=0.5, density=.6)
plt.axis('square')
plt.axis([-3, 3, -3, 3])
plt.show()
```

```python
#1b
def dF(x, y):
    return -3*x+2*y, x-2*y

x = np.arange(-3,4,1)
v_1 = eigen(x,-2,1)
v_2 = eigen(x,1,1)

X, Y = np.meshgrid(np.linspace(-3.0, 3.0, 30), np.linspace(-3.0, 3.0, 30))
dX, dY = dF(X, Y)
plt.figure(figsize=(15,15))
ax1 = plt.subplot(2,1,1)
plt.streamplot(X, Y, dX, dY, color='r', linewidth=0.5, density=.6)
plt.plot(x,v_1)
plt.plot(x,v_2)
plt.axis('square')
plt.axis([-3, 3, -3, 3])
plt.show()
```

```python
#1c
def dF(x, y):
    return -3*x+4*y, -2*x+3*y

v_1 = eigen(x,2,1)
v_2 = eigen(x,1,1)

X, Y = np.meshgrid(np.linspace(-3.0, 3.0, 30), np.linspace(-3.0, 3.0, 30))
dX, dY = dF(X, Y)

plt.streamplot(X, Y, dX, dY, color='r', linewidth=0.5, density=.6)
plt.plot(x,v_1)
plt.plot(x,v_2)
plt.axis('square')
plt.axis([-3, 3, -3, 3])
plt.show()
```
