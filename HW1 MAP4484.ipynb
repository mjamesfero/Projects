```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# Initialization
tstart = 0
tstop = 1
increment = .01
x0 = 1.95
t = np.arange(tstart,tstop,increment)
# Function that returns dx/dt
def mydiff(x, t):
    dxdt = 4 * x**2 - 16
    return dxdt
# Solve ODE
x = odeint(mydiff, x0, t)
#print(x)
# Plot the Results
plt.plot(t,x)
plt.title('Plotting Differential Equation Solution')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
```

```python
#Problem 2 a
# 1D arrays

y = np.arange(-3,3,0.1)

# Meshgrid
X,Y = np.meshgrid(x,y)
one = np.repeat(.5,len(X)).T

# Assign vector directions
Ex = np.concatenate(([one],[one],[one],[one],[one]),axis=0).T
#Ex =  X#(X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2)
Ey = 4 * Y**2 - 16

# Depict illustration
plt.figure#(figsize=(10, 10))
plt.streamplot(X,Y,Ex,Ey, density=1.4, linewidth=None, color='#A23BEC')
#plt.plot(-1,0,'-or')
#plt.plot(1,0,'-og')
plt.title('Vector Field')

# Show plot with grid
plt.grid()
plt.show()
```

```python
#Problem 2 b
# 1D arrays
x = np.arange(-5,5,1)
y = np.arange(-30,20,1)

# Meshgrid
X,Y = np.meshgrid(x,y)
one = np.repeat(.01,len(X)).T

# Assign vector directions
Ex = [one]
for ii in range(len(Y[0])-1):
    Ex = np.concatenate((Ex,[one]),axis=0)
Ex = Ex.T
Ey = (np.exp(Y) - np.cos(Y))/100

# Depict illustration
plt.figure#(figsize=(10, 10))
plt.streamplot(X,Y,Ex,Ey, density=1.5, linewidth=None, color='#A23BEC')
plt.title('Vector Field')

# Show plot with grid
plt.grid()
plt.show()
```
