1. [35 points] You are observing variable stars in the Small Magellanic Cloud. You see five stars with the following periods and apparent magnitudes. For each star (i) identify what type of variable it is, (ii) use the Ks-band period-luminosity relations listed below from Gaia Collaboration et al. (2017), A&A, 605, A79 and from Matsunaga et al. (2009), MNRAS, 399, 4 to estimate the distance modulus (m-M), and (iii) compute the distance estimated using each star. Note that the slide from class showing period-luminosity relations may be helpful.
#### Period-Luminosity Relations
Classical Cepehids (Type I): $M_{Ks} =-3.365\log(P)-2.6$
<br>
Type II Cepheids: $M_{Ks} =-2.385\log(P)-1.51$
<br>
RR Lyrae: $M_{Ks} =-2.73\log(P)-1.22$

Star 1: $m_{Ks} = 17.06$, Period = 4.36 days
<br>

```python
#star 1
#period of 4.36 days -> most likely a cepheid, probably type 2
period1 = 4.36
ab_mag1 = -2.385*np.log(period1)/np.log(10)-1.151
dmod1 = 17.06-ab_mag1
distance1 = 10**((dmod1+5)/5)
print(f'Absolute mag: {ab_mag1}, Distance modulus: {dmod1}, Distance: {distance1}pc')
```

2. [25 points] There remains a large debate about the functional form of the initial mass function, specifically with regard to whether it has the same form in all places (as we discussed in class). It has been argued in recent years that in massive elliptical galaxies the IMF is close to Salpeter, while in lower mass ellipticals it is closer to Kroupa. One reason people care is that this impacts the inferred total stellar mass in these galaxies. Including all stars above the brown dwarf limit, calculate the ratio of total stellar mass for a Salpeter IMF compared to a Kroupa IMF. This can be considered the systematic uncertainty when measuring the stellar mass of a galaxy. Assume that the number of stars with 1 solar mass is the same for both the Kroupa and Salpeter IMFs.
<br>
For Salpeter IMF, we have $\xi(M)=\xi_0M^{-2.35}$ and $\xi(1)=\xi_01^{-2.35}=\xi_0$. For Kroupa IMF, we have 
\[ \xi(M) = \begin{cases} 
          \xi_0M^{-1.3} & 0.08\leq M\leq 0.5 \\
          \xi_0M^{-2.7} &  0.5\leq M
       \end{cases}
    \]
and $\xi(1)=\xi_01^{-2.7}=\xi_0$. We can see that both IMFs have the same constant $\xi_0$. 
<br>
The mass integral for Salpeter is then $m(M)=\int_{0.08}^{\infty}M\xi(M)dM=\int_{0.08}^{\infty}\xi_0M^{-1.35}dM=\frac{\xi_0}{0.35}M^{-0.35}\big]_{\infty}^{0.08}=\xi_06.916$. The mass integrals for Kroupa are then $m(M)=\int_{0.08}^{\infty}M\xi(M)dM=\int_{0.08}^{0.5}\xi_0M^{-0.3}dM+\int_{0.5}^{\infty}\xi_0M^{-1.7}dM=\frac{\xi_0}{0.7}M^{0.7}\big]_{0.08}^{0.5}-\frac{\xi_0}{0.7}M^{-0.7}\big]_{0.5}^{\infty}=\xi_02.956$.
<br>
The ratio of Salpeter to Kroupa is $\dfrac{\xi_06.916}{\xi_02.956}=2.34$.
